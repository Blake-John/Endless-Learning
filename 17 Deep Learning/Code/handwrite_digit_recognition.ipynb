# How the data is in Numpy ?

Assume that we have a matrix : 

$$A = 
\begin{bmatrix}
1 && 2 && 3 \\
4 && 5 && 6
\end{bmatrix}
$$

So the matrix is $2 \times 3$ matrix, and can be represented as : 

```python
import numpy as np

x = np.array ([[1, 2, 3],
               [4, 5, 6]])
print (x)
```

If another matrix : 

$$B = 
\begin{bmatrix}
0.1 && 0.2 \\
-3 && -4 \\
-.5 && -.6 \\
7 && 8
\end{bmatrix}
$$

it is represented as :

```python
x_ = np.array ([[0.1, 0.2],
               [-3.0, -4.0],
               [-0.5, -.6],
               [7, 8]])
print (x_)
```

# Tensorflow Implementation

Now, we are going to implement the neural network by tensorflow :

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError
import matplotlib.pyplot as plt
import numpy as np
```

Then, we can build our model and check its structure :

```python
layer_1 = Dense (units=25, activation='relu')
layer_2 = Dense (units=15, activation='relu')
layer_3 = Dense (units=1)

model = Sequential ([
    layer_1, layer_2, layer_3
])

model.summary ()   # show the details of the model
```

```python
x = np.linspace (-5, 5, 100)
x = x.reshape (1, -1)
print (x.shape)

y = x ** 2 + 5 * np.random.rand (1, 100)
plt.scatter (x, y)
plt.show ()
```

```python
model.compile (loss=MeanSquaredError ())
model.fit (x, y, epochs=2000)
```

```python
model.summary ()
pred = model.predict (x)
plt.plot (x, pred)
plt.show ()
```

## 1.1 Linear Activation Function

$$g (z) = z$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace (-5, 5, 100)

def l_g (x) : 
    return x

y = l_g (x)

plt.plot (x, y)
plt.title ("Linear Acctivation Function", fontdict={"family":"Hack"})
plt.grid ()
plt.show ()
```

## 12 Sigmoid

$$g(z) = \dfrac{1}{1 + e^{-z}}, \ z = \vec{w} \cdot \vec{x} + b$$

```python
def sigmoid (x) : 
    return 1 / (1 + np.exp (-x))

y = sigmoid (x)

plt.plot (x, y)
plt.title ("Sigmoid", fontdict={"family":"Hack"})
plt.grid ()
plt.show ()
```

## 1.3 ReLU

$$g (z) = \max \{0, z\}$$ 

```python
def ReLU (x) : 
    zero_ = np.zeros (x.shape)
    return np.maximum (0, x)

y = ReLU (x)

plt.plot (x, y)
plt.title ("ReLU", fontdict={"family":"Hack"})
plt.grid ()
plt.show ()
```

# Crossentropy

$$loss = -\log a_j \ \ \ \ if\ y = j$$

```python
x = np.linspace (0, 1.2, 50)
y = -np.log (x)

plt.plot (x, y)
plt.title ("Crossentropy Loss", fontdict={"family":"Hack"})
plt.grid ()
plt.show ()
```
