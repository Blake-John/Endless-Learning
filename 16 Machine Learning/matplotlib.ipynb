# Data Visualization - Matplotlib

Matplotlib is a low-level library of Python which is used for data visualization. It is easy to use and **emulates MATLAB like graphs and visualization**. This library is built on the top of NumPy arrays and consist of several plots like line chart, bar chart, histogram, etc. It provides a lot of flexibility but at the cost of writing more code.

Pyplot is a Matplotlib module that provides a MATLAB-like interface. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python and the advantage of being free and open-source. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. The various plots we can utilize using Pyplot are **Line Plot**, **Histogram**, **Scatter**, **3D Plot**, **Image**, **Contour**, and **Polar**.

## 01 Create a Graph

### 1.1 Plot the data

In python, we can plot **normal lists or numpy arrays in the graph by pyplot** :

```python
import numpy as np
import matplotlib.pyplot as plt

# get the numpy array
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot (x, y)
"""
(function) def plot(
    *args: ArrayLike,
    scalex: bool = True,
    scaley: bool = True,
    data: Any | None = None,
    **kwargs: Any
) -> list[Line2D]
"""

# show the graph
plt.show ()
```

### 1.2 Adding Title

The `plt.title()` method in matplotlib module is used to **specify the title of the visualization depicted** and displays the title using various attributes : 

```python
plt.plot (x, y)

# setting the title
plt.title ("Linear Graph")
"""
(function) def title(
    label: str,
    fontdict: dict[str, Any] | None = None,
    loc: Literal['left', 'center', 'right'] | None = None,
    pad: float | None = None,
    *,
    y: float | None = None,
    **kwargs: Any
) -> Text
"""
plt.show ()
```

We can also change the appearance of the title by using the parameters of this function.

```python
plt.plot (x, y)
plt.title ("Linear Graph", 
           fontdict={"family": "Hack",
                     "style": "italic",
                     "weight": "bold",
                     "size": 16,
                     "color": "#5EB9CB"},
           loc="center",
           )
"""
family - 'Hack', 'monospace' ...
style - 'normal', 'italic', 'oblique'。
variant - 'normal', 'small-caps'。
weight - 'normal', 'bold', 'light', 'ultrabold', 'heavy' ...
size - 12, 14, 16 ...
stretch - the horizontal stretch like
    'normal', 'condensed', 'semi-condensed', 'expanded', 'semi-expanded'。
bbox - the dictionary include the property of the title bbox
color - 'black', 'red', '#008000', '0.8' ...
"""
plt.show ()
```

### 1.2 Axis Label

In layman’s terms, the X label and the Y label are the titles given to X-axis and Y-axis respectively. These can be added to the graph by using the `plt.xlabel()` and `plt.ylabel()` methods : 

```python
plt.plot (x, y)
plt.title ("Linear Graph", fontdict={"family": "Hack"})

# add the x, y label
plt.xlabel ("X-Axis")
"""
(function) def xlabel(
    xlabel: str,
    fontdict: dict[str, Any] | None = None,
    labelpad: float | None = None,
    *,
    loc: Literal['left', 'center', 'right'] | None = None,
    **kwargs: Any
) -> Text
"""

plt.ylabel ("Y-Axis")
"""
(function) def ylabel(
    ylabel: str,
    fontdict: dict[str, Any] | None = None,
    labelpad: float | None = None,
    *,
    loc: Literal['bottom', 'center', 'top'] | None = None,
    **kwargs: Any
) -> Text
"""

plt.show ()
```

### 1.3 Setting Limits and Tick labels

You might have seen that Matplotlib **automatically sets the values and the markers(points)** of the X and Y axis, however, it is possible to set the limit and markers manually. `plt.xlim()` and `plt.ylim()` functions are used to **set the limits** of the X-axis and Y-axis respectively. Similarly, `plt.xticks()` and `plt.yticks()` functions are used to **set tick labels** : 

```python
plt.plot (x, y)
plt.title ("Linear Graph", fontdict={"family": "Hack"})
plt.xlabel ("X-Axis")
plt.ylabel ("Y-Axis")

# set the limit of y-axis
plt.ylim (0, 80)
"""
(function) def ylim(
    *args: Any - always (bottom, top),
    **kwargs: Any
) -> tuple[float, float]
"""

# set the ticks of x-axis
plt.xticks (x, labels=["one", "two", "three", "four"])
"""
(function) def xticks(
    ticks: ArrayLike | None = None,
    labels: Sequence[str] | None = None,
    *,
    minor: bool = False,
    **kwargs: Any
) -> tuple[list[Tick] | ndarray, list[Text]]
"""

plt.show ()
```

### 1.4 Adding Legends

A legend is **an area describing the elements of the graph**. In simple terms, it **reflects the data displayed in the graph’s Y-axis**. It generally appears **as the box containing a small sample** of each color on the graph and a small description of what this data means :

```python
plt.plot (x, y)
plt.title ("Linear Graph", fontdict={"family": "Hack"})
plt.xlabel ("X-Axis")
plt.ylabel ("Y-Axis")
plt.ylim (0, 80)
plt.xticks (x, labels=["one", "two", "three", "four"])

# set the legend
plt.legend (["Legend 1"])
"""
(function) def legend(
    *args: Any,
    **kwargs: Any
) -> Legend
"""

plt.show ()
```

If we use two group of data, it will have a better insight into `plt.legend ()` : 

```python
x1 = [10, 20, 30, 40, 50]
y1 = [10, 15, 20, 25, 30]

x2 = [10, 20, 30, 40, 50]
y2 = [20, 22, 24, 26, 28]

# add the label when plot the data
plt.plot (x1, y1, label="Series 1")
plt.plot (x2, y2, label="Series 2")

plt.title ("Linear Graph", fontdict={"family": "Hack"})
plt.xlabel ("X-Axis")
plt.ylabel ("Y-Axis")

plt.legend ()
plt.show ()
```

```python
x1 = [10, 20, 30, 40, 50]
y1 = [10, 15, 20, 25, 30]

x2 = [10, 20, 30, 40, 50]
y2 = [20, 22, 24, 26, 28]

plt.plot (x1, y1, label="Series 1")
plt.plot (x2, y2, label="Series 2")

plt.title ("Linear Graph", fontdict={"family": "Hack"})
plt.xlabel ("X-Axis")
plt.ylabel ("Y-Axis")

# change the legend label when setting the legend
plt.legend (["Legend 1", "Legend 2"], 
            loc="lower right",
            shadow=True,
            fontsize='medium',
            ncol=2,
            title="Legend Title")
"""
loc - the position of legend, can be 'best'
    'lower/upper left/center/right'
shadow - make the legend shadowing
fontsize - can be string 'large', 'small', or int
ncol - the columns of the legend
title - the title of the legend
"""

plt.show ()
```

## 1.5 Adding Grid

We can add grid to the background : 

```python
x1 = [10, 20, 30, 40, 50]
y1 = [10, 15, 20, 25, 30]

x2 = [10, 20, 30, 40, 50]
y2 = [20, 22, 24, 26, 28]

plt.plot (x1, y1, label="Series 1")
plt.plot (x2, y2, label="Series 2")

plt.title ("Linear Graph", fontdict={"family": "Hack"})
plt.xlabel ("X-Axis")
plt.ylabel ("Y-Axis")

# change the legend label when setting the legend
plt.legend (["Legend 1", "Legend 2"], 
            loc="lower right",
            shadow=False,
            fontsize='medium',
            ncol=2,
            title="Legend Title")
plt.grid (True)
plt.show ()
```

## 02 Classes of Matplotlib

There are two important classes in matplotlib

* Figure
* Axes

### 2.1 Figure Class

Consider the figure class as the **overall window or page** on which everything is drawn. It is a **top-level container that contains one or more axes**. A figure can be created using the `plt.figure()` method.

```python
# Creating a new figure with 
# width = 7 inches
# height = 5 inches 
# face color as green
# edgecolor as red

fig = plt.figure (num=2,
                  figsize=(7, 5),
                  facecolor='g',
                  edgecolor='b')

"""
(function) def figure(
    num: int | str | Figure | SubFigure | None = None,
    figsize: tuple[float, float] | None = None,
    dpi: float | None = None,
    *,
    facecolor: ColorType | None = None,
    edgecolor: ColorType | None = None,
    frameon: bool = True,
    FigureClass: type[Figure] = Figure,
    clear: bool = False,
    **kwargs: Any
"""

# Creating a new axes for the figure
ax1 = fig.add_axes([1, 1, 1, 1])
ax2 = fig.add_axes([2, 2, 2, 2])

# Adding the data to be plotted
ax1.plot (x1, y1)
ax2.plot (x2, y2)

plt.show ()
```

### 2.2 Axes Class

Axes class is the most basic and flexible unit for **creating sub-plots**. A given figure may **contain many axes**, but a given axes can only be present in one figure. The `plt.axes()` function creates the axes object. 

Just like pyplot class, axes class also provides methods for adding titles, legends, limits, labels, etc. Let’s see a few of them

* Adding Title – `ax.set_title()`
* Adding X Label and Y label – `ax.set_xlabel()`, `ax.set_ylabel()`
* Setting Limits – `ax.set_xlim()`, `ax.set_ylim()`
* Tick labels – `ax.set_xticklabels()`, `ax.set_yticklabels()`
* Adding Legends – `ax.legend()`

```python
fig = plt.figure(figsize = (5, 4))
 
# Adding the axes to the figure
ax = fig.add_axes([1, 1, 1, 1])
 
# plotting 1st dataset to the figure
ax1 = ax.plot(x, y)
 
# plotting 2nd dataset to the figure
ax2 = ax.plot(y, x)
 
# Setting Title
ax.set_title("Linear Graph")
 
# Setting Label
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
 
# Adding Legend
ax.legend(labels = ('line 1', 'line 2'))
 
plt.show()
```

## 03 Mutiple Plot

We have learned about the basic components of a graph that can be added so that it can convey more information. One method can be by calling the plot function again and again with a different set of values as shown in the above example. Now let’s see how to **plot multiple graphs** using some functions and also how to plot subplots. 

### 3.1 `add_axes ()`

The `add_axes()` method is used to add axes to the figure. This is a method of **figure class**


```python
fig = plt.figure (figsize=(5, 4))

# create first axes for the figure
ax1 = fig.add_axes ([0.1, 0.1, 0.8, 0.8])
"""
(method)
def add_axes(ax: Axes) -> Axes: ...

def add_axes(
    rect: tuple[left, bottom, width, height],
    projection: str | None = ...,
    polar: bool = ...,
    **kwargs: Any
) -> Axes: ...
"""

# create second axes for the figure
ax2 = fig.add_axes ([1, 0.1, 0.8, 0.8])

ax1.plot (x, y)
ax2.plot (y, x)

plt.show ()
```

### 3.2 `subplot ()`

This method adds another plot at the specified grid position in the current figure.

```python
fig = plt.figure ()

# adding first subplot
plt.subplot (1, 2, 1)
"""
subplot(nrows, ncols, index, **kwargs)
subplot(pos, **kwargs)
subplot(ax)

pos - the int represent nrows, ncols, index
    such as 121 - 1 row, 2 cols, the first subplot
"""

plt.plot (x, y)

# adding another subplot
plt.subplot (1, 2, 2)
plt.plot (y, x)

plt.show ()
```

```python
fig = plt.figure ()

# adding first subplot
plt.subplot (2, 1, 1)
"""
subplot(nrows, ncols, index, **kwargs)
subplot(pos, **kwargs)
subplot(ax)

pos - the int represent nrows, ncols, index
    such as 121 - 1 row, 2 cols, the first subplot
"""

plt.plot (x, y)

# adding another subplot
plt.subplot (2, 1, 2)
plt.plot (y, x)

plt.show ()
```

```python
# another way to use subplot is
fig, axes = plt.subplots (1, 2)

axes[0].plot (x, y)
axes[0].plot (y, x)

axes[1].plot (x, y)

plt.show ()
```

## 04 Different Types of Plot

Matplotlib supports **a variety of plots** including **line charts**, **bar charts**, **histograms**, **scatter plots**, etc. We will discuss the most commonly used charts in this article with the help of some good examples and will also see how to customize each plot.  

### 4.1 Line Chart

Line chart is one of the basic plots and can be created using the `plt.plot()` function. It is used to represent a relationship between two data X and Y on a different axis.

```python
x = [10, 20, 30, 40, 50]
y = [15, 20, 30, 35, 50]

plt.subplot (221)
plt.plot (x, y, marker="x", linestyle="-")
"""
(function) def plot(
    *args: ArrayLike,
    scalex: bool = True,
    scaley: bool = True,
    data: Any | None = None,
    **kwargs: Any
) -> list[Line2D]
"""
"""**kwargs :
color -         Change the color of the line
linewidth -     Customizing the width of the line
marker -        For changing the style of plotted point
    available : "." "o" "," "x" "v" "^" "<" ">" "s" "p" "*" "+"
                "1" "2" "3" "4" "h" "H" "d" "D" "|" "_"
markersize -    For changing the size of the markers
linestyle -     For defining the style of the plotted line
    available : "-" "--" "-." ":"

"""

plt.subplot (222)
plt.plot (x, y, linestyle="--", marker=".")

plt.subplot (223)
plt.plot (x, y, linestyle="-.", marker="D")

plt.subplot (224)
plt.plot (x, y, linestyle=":", marker="s")


plt.show ()
```

### 4.2 Bar Chart

A bar chart is a graph that **represents the category of data with rectangular bars** with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted **horizontally or vertically**. A bar chart describes the comparisons between the discrete categories. It can be created using the `plt.bar()` method.


```python
x = ["Sun", "Mon", "Tue", "Wed", "Thur", "Fri", "Sat"]
y = [40.9, 50.2, 48.3, 46.6, 45.7, 49.2, 47]

plt.bar (x, y)
"""
(function) def bar(
    x: ArrayLike,
    height: ArrayLike,
    width: ArrayLike = 0.8,
    bottom: ArrayLike | None = None,
    *,
    align: Literal['center', 'edge'] = "center",
    data: Any | None = None,
    **kwargs: Any
"""
"""**kwargs : 
color -     Bar faces
edgecolor - Color of edges of the bar
"""

plt.xlabel ("Day")
plt.ylabel ("Bill")

plt.show ()
```

### 4.3 Histogram

A histogram is basically used to represent **data provided in a form of some groups**. It is a type of bar plot where the X-axis represents the **bin ranges** while the Y-axis **gives information about frequency**. The `plt.hist()` function is used to compute and create histogram of x.

```python
x = [9, 10, 12, 15, 23, 24, 25, 26, 30, 33, 36, 35, 39, 31, 37, 40, 45, 46, 41, 51, 50, 52, 61]

plt.hist (x, bins=10)
"""
(function) def hist(
    x: ArrayLike | Sequence[ArrayLike],
    bins: int | Sequence[float] | str | None = None,
    range: tuple[float, float] | None = None,
    density: bool = False,
    weights: ArrayLike | None = None,
    cumulative: bool | float = False,
    bottom: ArrayLike | None = None,
    histtype: Literal['bar', 'barstacked', 'step', 'stepfilled'] = "bar",
    align: Literal['left', 'mid', 'right'] = "mid",
    orientation: Literal['vertical', 'horizontal'] = "vertical",
    rwidth: float | None = None,
    log: bool = False,
    color: ColorType | Sequence[ColorType] | None = None,
    label: str | Sequence[str] | None = None,
    stacked: bool = False,
    *,
    data: Any | None = None,
    **kwargs: Any
) -> tuple[ndarray | list[ndarray], ndarray, BarContainer | Polygon | list[BarContainer | Polygon]]
"""
"""
bins -      Number of equal-width bins 
color -     For changing the face color
edgecolor - Color of the edges
linestyle - For the edgelines
alpha -     blending value, between 0 (transparent) and 1 (opaque)
"""

plt.show ()
```

### 4.4 Scatter Plot

Scatter plots are used to **observe relationships between variables**. The `plt.scatter()` method in the matplotlib library is used to draw a scatter plot.

```python
x = [9, 10, 12, 15, 23, 24, 25, 26, 30, 33, 36, 35, 39, 31, 37, 40, 45, 46, 41, 51, 50, 52, 61]
y = [10, 12, 14, 17, 22, 23, 27, 25, 29, 30, 32, 36, 37, 39, 35, 39, 40, 43, 45, 47, 44, 42, 50]

plt.scatter (x, y, c="g", marker="x")
"""
(function) def scatter(
    x: ArrayLike,
    y: ArrayLike,
    s: ArrayLike | None = None,
    c: ArrayLike | ColorType | Sequence[ColorType] | None = None,
    marker: MarkerType | None = None,
    cmap: str | Colormap | None = None,
    norm: str | Normalize | None = None,
    vmin: float | None = None,
    vmax: float | None = None,
    alpha: float | None = None,
    linewidths: float | Sequence[float] | None = None,
    *,
    edgecolors: ColorType | Sequence[ColorType] | None = None,
    plotnonfinite: bool = False,
    data: Any | None = None,
    **kwargs: Any
) -> PathCollection
"""

plt.show ()
```

### 4.5 Pie Chart

Pie chart is a circular chart used to display only one series of data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges. It can be created using the `plt.pie()` method.

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

plt.pie (data, labels=cars)
"""
(function) def pie(
    x: ArrayLike,
    explode: ArrayLike | None = None,
    labels: Sequence[str] | None = None,
    colors: ColorType | Sequence[ColorType] | None = None,
    autopct: str | ((float) -> str) | None = None,
    pctdistance: float = 0.6,
    shadow: bool = False,
    labeldistance: float | None = 1.1,
    startangle: float = 0,
    radius: float = 1,
    counterclock: bool = True,
    wedgeprops: dict[str, Any] | None = None,
    textprops: dict[str, Any] | None = None,
    center: tuple[float, float] = (0, 0),
    frame: bool = False,
    rotatelabels: bool = False,
    *,
    normalize: bool = True,
    hatch: str | Sequence[str] | None = None,
    data: Any | None = None
) -> (tuple[list[Wedge], list[Text]] | tuple[list[Wedge], list[Text], list[Text]])
"""

plt.show ()
```

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR',]
data = [23, 10, 35, 15, 12]

plt.pie (data, labels=cars,
         explode=[0.1, 0.2, 0, 0, 0],
         autopct="%.2f%%")

plt.show ()
```
